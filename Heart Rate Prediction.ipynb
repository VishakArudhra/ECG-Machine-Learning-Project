#### Frequency Domain Features:

**absolute power** : It's the integral of all the power values within the frequency range of a band. 

**relative power** : Absolute power of a band as the percentage of the total absolute power of all bands. 

The four frequency bands are ultra low frequency, very low frequency, low frequency and high frequency. Read more on the **four domain analysis using Fast Fourier Transform.**

##### Some important definitions:

* The HF or respiratory band (0.15‚Äì0.40‚ÄâHz) is influenced by breathing from 9 to 24‚Äâbpm. 


*  The LF band (0.04‚Äì0.15‚ÄâHz) is comprised of rhythms with periods between 7 and 25‚Äâs and is affected by breathing from ~3 to 9‚Äâbpm.


* The ratio of **LF to HF power (LF/HF ratio)** may estimate the ratio between **sympathetic nervous system (SNS)** and **parasympathetic nervous system (PNS)** activity under controlled conditions.
